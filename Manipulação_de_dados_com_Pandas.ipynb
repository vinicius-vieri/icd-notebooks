Pandas é uma biblioteca de código aberto do Python usada para análise e manipulação de dados. Ela fornece ferramentas de análise de dados fáceis de usar e de alto desempenho, tornando-se uma das bibliotecas mais populares para trabalhar com dados no Python.

A biblioteca Pandas oferece duas principais estruturas de dados: Series e DataFrame. A Series é uma matriz unidimensional rotulada capaz de armazenar qualquer tipo de dados, como inteiros, strings, objetos Python, etc. O DataFrame é uma estrutura de dados bidimensional rotulada, semelhante a uma tabela de banco de dados, onde cada coluna pode ter um tipo de dados diferente.

Algumas das principais funcionalidades do Pandas incluem:

Leitura e gravação de arquivos em diversos formatos, como CSV, Excel, SQL, JSON, etc.
Manipulação de dados, como seleção, filtragem, limpeza, transformação, agrupamento e agregação.
Tratamento de dados ausentes (missing values).
Cálculo de estatísticas descritivas, como média, desvio padrão, correlação, etc.
Visualização de dados com gráficos.
O Pandas é frequentemente usado em conjunto com outras bibliotecas do Python, como NumPy, Matplotlib e Scikit-Learn, para análise de dados e aprendizado de máquina.

# Query

Em pandas, uma query é uma forma de filtrar dados em um DataFrame baseado em uma expressão booleana.

A query funciona passando uma string de consulta para o método query() de um DataFrame. Essa string é uma expressão booleana que avalia cada linha do DataFrame e retorna True ou False. As linhas que retornam True são mantidas no DataFrame, enquanto as que retornam False são descartadas.

```python
# vamos descrever os dados do dataset anterior

df.describe()
```

repare que ele já realiza algumas operações de estatística descritiva.

# Consultas por operador lógico ou não

Em consultas pandas, os operadores lógicos são usados ​​para combinar várias condições em uma expressão booleana que pode ser avaliada para cada linha do DataFrame. Existem três operadores lógicos principais em pandas: and, or e not.

O operador and é usado para combinar duas ou mais condições e retornar True somente se todas as condições forem True. 

O operador or é usado para combinar duas ou mais condições e retornar True se pelo menos uma das condições for True. 

O operador not é usado para negar uma condição e retornar True se a condição for falsa. 

Além disso, também é possível usar parênteses para agrupar as condições e controlar a ordem de avaliação das operações lógicas.

Os operadores lógicos são ferramentas poderosas para filtrar e selecionar dados em um DataFrame com base em múltiplas condições. Combinados com operadores de comparação, como <, >, <= e >=, eles permitem a construção de expressões booleanas complexas para consultas pandas.


```python
# gostaríamos de saber as vendas coporativas na região Central

df[ (df.Segmento == "Corporate") & (df.Regiao == "Central")].head() # usamos operador and, mas poderia ser ou e not
```

```python
# agrupando os dados com goupby

df[["Segmento", "Regiao", "Valor_Venda"]].groupby(["Segmento", "Regiao"]).mean()
```

```python
# podemos usar o método agg para agregar de maneira múltipla, pedindo mais de um tipo de saída

df[["Segmento", "Regiao", "Valor_Venda"]].groupby(["Segmento", "Regiao"]).agg(["mean", "median", "std"])
```

```python
# Filtrando uma coluna pela string "Corporate"

df[df.Segmento.str.startswith("Corporate")]
```

```python
df.Segmento.value_counts()
```

Em pandas, é possível construir gráficos a partir de um DataFrame usando a biblioteca de visualização de dados matplotlib. A matplotlib é uma biblioteca de plotagem em Python que fornece uma ampla gama de ferramentas para criar gráficos de alta qualidade.

Para criar um gráfico a partir de um DataFrame, primeiro você precisa importar as bibliotecas pandas e matplotlib e carregar seus dados em um DataFrame. Em seguida, use o método plot() do DataFrame para criar o gráfico.

```python
import sklearn
from sklearn.datasets import load_iris
data = load_iris()
```

```python
df = pd.DataFrame(data["data"], columns = data["feature_names"])
df["species"] = data["target"]
```

```python
df.plot()
```

```python
df.groupby('species').mean().plot.bar()
```

```python
columns = ['sepal length (cm)', 'petal length (cm)', 'petal width (cm)', 'sepal width (cm)']
df[columns].plot.box()
```
