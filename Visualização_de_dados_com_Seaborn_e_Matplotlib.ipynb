# Introdução 

Matplotlib é uma biblioteca de visualização de dados em Python que permite a criação de gráficos 2D de alta qualidade. Com ela, é possível criar gráficos de linha, de barra, de dispersão, histogramas, gráficos de contorno, entre outros tipos de gráficos. É uma das bibliotecas mais populares para visualização de dados em Python, sendo muito utilizada em áreas como ciência de dados, estatística, engenharia, finanças, entre outras.

Seaborn é uma biblioteca de visualização de dados em Python que é construída em cima do Matplotlib. Ela fornece uma interface de alto nível para criação de gráficos estatísticos mais complexos e atraentes visualmente. Seaborn também inclui recursos adicionais, como suporte para visualização de dados categóricos, gráficos de distribuição, gráficos de regressão, gráficos de matriz de correlação, entre outros. Seaborn é frequentemente utilizada em análises exploratórias de dados e em apresentações de dados em relatórios e artigos científicos.

Link oficial matplotlib: https://matplotlib.org/

Link oficial seaborn: https://seaborn.pydata.org/

```python
# Gráfico de linha
# Para usar métodos personalizados como inspace, precisamos importar o pylab
from pylab import *
%matplotlib inline
```

```python
# Dados
x = linspace(0, 5, 10)
y = x ** 3

# criando a figura
fig = plt.figure()

# definindo a escala dos eixos
axes = fig.add_axes([0, 0, 0.8, 0.8])

# plotando
axes.plot(x, y, "r")

# Labels e título
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Modelo")
```

```python
# Gráficos de linha duplo


x = linspace(0, 5, 20)
y = x ** 4

fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # eixos da figura principal
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # eixos da figura secundária

# Figura principal
ax1.plot(x, y, "r")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("Modelo primário")

# Figura secundária
ax2.plot(y, x, "g")
ax2.set_xlabel("y")
ax2.set_ylabel("x")
ax2.set_title("Modelo secundário");
```

```python
# modelo de gráfico com grid

x = linspace(0, 5, 10)
y = x ** 2

fig, axes = plt.subplots(1, 2, figsize = (10,3))

# Grid padrão
axes[0].plot(x, x**2, x, x**3, lw = 2)
axes[0].grid(True)

# Grid customizado
axes[1].plot(x, x**2, x, x**3, lw = 2)
axes[1].grid(color = 'b', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)
```

```python
# Diferentes estilos de plotagem 

# Dados
xx = np.linspace(-0.75, 1., 100)
n = np.array([0,1,2,3,4,5])

# Subplots
fig, axes = plt.subplots(1, 4, figsize = (12,3))

# 1
axes[0].scatter(xx, xx + 0.25 * randn(len(xx)), color = "b")
axes[0].set_title("Modelo de dispersão")

# 2
axes[1].step(n, n ** 2, lw = 2, color = "k")
axes[1].set_title("Modelo de etapas")

# 3
axes[2].bar(n, n ** 2, align = "center", width = 0.7, alpha = 0.7, color = "m")
axes[2].set_title("Modelo de barras")

# 4
axes[3].fill_between(x, x ** 2, x ** 3, alpha = 0.7, color = "g");
axes[3].set_title("Modelo de preenchimento");
```

```python
# Modelo de histograma

import random

# Dados
n = np.random.randn(500000)

# Cria os subplots
fig, axes = plt.subplots(1, 2, figsize=(12,7))

# 1
axes[0].hist(n)
axes[0].set_title("Histograma")
axes[0].set_xlim((min(n), max(n)))

# 2
axes[1].hist(n, cumulative = True, bins = 50)
axes[1].set_title("Histograma Cumulativo")
axes[1].set_xlim((min(n), max(n)));
```

```python
# Modelos de gráfico 3D


from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
# Dados
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

# Função para um mapa de cores
def ColorMap(phi_m, phi_p):
    return ( + alpha - 2 * np.cos(phi_p)*cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p))

# Mais dados
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = ColorMap(X, Y).T

# Cria a figura
fig = plt.figure(figsize = (14,6))

# Adiciona o subplot 1 3d
ax = fig.add_subplot(1, 2, 1, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# Adiciona o subplot 2 3d
ax = fig.add_subplot(1, 2, 2, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Cria a barra de cores como legenda
cb = fig.colorbar(p, shrink=0.5)
```

# Seaborn

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib as mat
import warnings
import seaborn as sns
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
# no próprio seaborn existem datasets de teste

dados = sns.load_dataset("tips")
dados.head()
```

```python
# O método lmplot() cria plot com dados e modelos de regressão

sns.lmplot(data = dados, x = "total_bill", y = "tip", col = "smoker")
```

```python
# agora, vamos criar um dataframe pandas e alimentá-lo com dados aleatórios

df = pd.DataFrame()
df["Idade"] = random.sample(range(18, 90), 30)
df['peso'] = random.sample(range(65, 150), 30)
df.head()
```

```python
# lmplot
sns.lmplot(data = df, x = "Idade", y = "peso", fit_reg = True)
```

```python
# kdeplot
sns.kdeplot(df.Idade)

# kdeplot
sns.kdeplot(df.peso)
```

```python
# distplot

sns.distplot(df.Idade)

sns.distplot(df.peso)
```

```python
# Histograma
plt.hist(df.Idade, alpha = .4)
sns.rugplot(df.Idade)

plt.hist(df.peso, alpha = .4)
sns.rugplot(df.peso)
```

```python
# Box Plot
sns.boxplot(df.Idade, color = 'b')
```

```python
# Box Plot
sns.boxplot(df.peso, color = 'k')
```

```python
# Clustermap

# Um clustermap é um tipo de gráfico hierárquico que visualiza as relações entre as linhas e as colunas de uma matriz de dados. 

sns.clustermap(df)
```

```python
# Por último, vamos usar os conhecimentos para visualizar um gráfico estatístico na questão dos fumantes.


# Valores randômicos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

# Variáveis
flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

# Dataframe
dados = pd.DataFrame({"altura": altura, "peso": peso, "flag_fumante": flag_fumante})

# Cria os dados para a variável flag_fumante
dados["flag_fumante"] = dados["flag_fumante"].map({True: "Fumante", False: "Não Fumante"})
```

```python
# Style
sns.set(style = "ticks")

# lmplot
sns.lmplot(x = "altura", 
           y = "peso", 
           data = dados, 
           hue = "flag_fumante", 
           palette = ["tab:blue", "tab:red"], 
           height = 7)

# Labels e título
plt.xlabel("Altura")
plt.ylabel("Peso")
plt.title("Relação entre altura e peso de fumantes e não fumantes")

# Show
plt.show()
```
